# RenAIssance: Handwritten Text Recognition for Early Modern Spanish

This notebook demonstrates an end-to-end HTR pipeline using a Vision-Language Model (VLM) approach (Gemini 1.5). 

### Objectives:
1.  **VLM-based Transcription**: Directly reading handwritten images.
2.  **Linguistic Correction**: Using LLM logic to refine archaic orthography.
3.  **Accuracy Evaluation**: Measuring CER and WER against ground truth.

```python
import os
import sys
from IPython.display import Image, display
from dotenv import load_dotenv

# Add src to path
sys.path.append(os.path.abspath('../src'))

from pipeline.vlm_bridge import VLMPipeline
from utils.metrics import calculate_cer, calculate_wer

load_dotenv()
print("Environment initialized.")
```

## 1. Load Sample Image
We use a sample manuscript image downloaded during setup.

```python
sample_path = "../data/raw/sample_manuscript.jpg"
if os.path.exists(sample_path):
    display(Image(filename=sample_path, width=600))
else:
    print("Sample image not found. Please place an image in data/raw/")
```

## 2. Run VLM Transcription
The model (Gemini 1.5 Flash) is prompted as an expert paleographer.

```python
try:
    pipeline = VLMPipeline()
    raw_transcription = pipeline.transcribe_image(sample_path)
    print("--- Raw Transcription ---")
    print(raw_transcription)
except Exception as e:
    print(f"Transcription error: {e}\nNote: Ensure GEMINI_API_KEY is set in .env")
```

## 3. Post-Processing & Correction
The LLM performs a second pass to correct obvious errors and handle archaic Spanish context.

```python
try:
    refined_text = pipeline.correct_text(raw_transcription)
    print("--- Refined Text ---")
    print(refined_text)
except Exception as e:
    print(f"Correction error: {e}")
```

## 4. Evaluation
Assuming we have ground truth for this segment.

```python
# Example Ground Truth (for demonstration)
ground_truth = refined_text # Placeholder for actual GT

cer = calculate_cer(ground_truth, refined_text)
wer = calculate_wer(ground_truth, refined_text)

print(f"CER: {cer:.4f}")
print(f"WER: {wer:.4f}")
```
